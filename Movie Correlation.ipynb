```python
#First step: Importing the libraries: Pandas, Seaborn, Numpy and Matplotlib
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Importing the dataset.csv file
df = pd.read_csv(r'C:\Users\Juan\Desktop\DS\MovieDS\movies.csv')
```

```python
#This code is for visualizing the headers of the table
df.head()
```

```python
#Here I use a for loop to look for the percentage of Nulls in each columns
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#This line is to check the type of data
df.dtypes
```

```python
#I used this line to turn Budget and Gross to Int64 and then I had to change it back to Float64 again
df['budget'] = df['budget'].astype('float64')
df['gross'] = df['gross'].astype('float64')
```

```python
#Check the dataframe again
df.head()
```

```python
#The year was strangely different to the released year, so I created the column correct_year
df['correct_year'] = df['released'].astype(str).str[:4]
```

```python
#This line is to sort the table by gross earning from highest to lowest
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#This is a Pandas setting to change the maximum number of displayed rows, by setting None, I am displaying all the rows by default
pd.set_option('display.max_rows', None)
```

```python
df.drop_duplicates()
```

```python
#I had to drop the rows with Null values in 'budget' and 'gross' because the amount was significant and the next steps would not work without this.
df = df.dropna(subset=['budget'])
df = df.dropna(subset=['gross'])
```

```python
#Check again for percentage of nulls, which is now 0 in budget and gross
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Creating a scatterplot with matplotlib between budget as x-axis and gross earnings as y-axis
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Film Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#Creating a scatter plot with a linear regression line with Seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#Checking for correlations to see if there is a high correlation between budget and gross by using Pearson correlation
df.corr(method='pearson')
```

```python
#We create a heatmap using the correlations table os the previous line
#This visualization shows the strongest correlations in a bright color and the weakest in a darker color
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#This line loops over every column and turn the ones that are 'Object' type into category type
#Then it assigns them an integer starting from 0 with cat.codes
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
    
df_numerized.head()
```

```python
#Now we can visualize more columns in this correlation heatmap 
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
#The 1s don't count because they are correlating with themselves
```

```python
#Correlations table of the numerized data frame
df_numerized.corr()
```

```python
#We create a correlations matrix
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
#Sorting them in pairs
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#Here I am going to see which sorted pairs have a high correlation, defining it as greater than 0.5
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
